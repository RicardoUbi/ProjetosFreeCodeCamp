```
# get data files
!wget https://cdn.freecodecamp.org/project-data/sms/train-data.tsv
!wget https://cdn.freecodecamp.org/project-data/sms/valid-data.tsv

train_file_path = "train-data.tsv"
test_file_path = "valid-data.tsv"
```

Dividindo os dados em treinamento e teste.

```
# Carregar dados
train_data = pd.read_csv(train_file_path, sep='\t', header=None)
test_data = pd.read_csv(test_file_path, sep='\t', header=None)

X_train = train_data[1]
y_train = train_data[0]
X_test = test_data[1]
y_test = test_data[0]
```

Vetorizar os textos: Transformamdo as mensagens de texto em representações numéricas que podem ser usadas por algoritmos de aprendizado de máquina. Neste caso, é utilizada a vetorização baseada na contagem de palavras, onde cada palavra única é atribuída a um índice inteiro e a contagem de ocorrências de cada palavra é usada como recurso.

```
# Vetorizar os textos
vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

Treinar o classificador Naive Bayes com as features (mensagens vetorizadas) e os labels de treinamento.

```
# Treinar o classificador Naive Bayes
classifier = MultinomialNB()
classifier.fit(X_train_vec, y_train)
```


Nea função, o texto fornecido como entrada é primeiro vetorizado usando o mesmo objeto vectorizer que foi ajustado aos dados de treinamento. Em seguida, o classificador Naive Bayes classifier é usado para prever a probabilidade de ser spam ou ham, retornamdo uma lista contendo a probabilidade e a classificação mais provável.

```
# Função para prever mensagem
def predict_message(pred_text):
    message_vec = vectorizer.transform([pred_text])
    prediction = classifier.predict_proba(message_vec)
    spam_probability = prediction[0][1]
    ham_probability = prediction[0][0]
    if spam_probability > ham_probability:
        return [spam_probability, 'spam']
    else:
        return [ham_probability, 'ham']

pred_text = "how are you doing today?"

prediction = predict_message(pred_text)
print(prediction)
```
