### Jogo do diagrama de dispersão
- Gera pontos aleatoriamente em um gráfico e o jogador deve inserir as coordenadas (x,y)
- Para aumentar a dificuldade, aumente o gráfico



```python
import matplotlib.pyplot as plt
import random

score = 0

xmin = -25
xmax = 25
ymin = -25
ymax = 25

fig, ax = plt.subplots()

for i in range(0,3):
    xpoint = random.randint(xmin, xmax)
    ypoint = random.randint(ymin, ymax)
    x = [xpoint]
    y = [ypoint]
    plt.axis([xmin,xmax,ymin,ymax]) # window size
    plt.plot([xmin,xmax],[0,0],'b') # blue x axis
    plt.plot([0,0],[ymin,ymax], 'b') # blue y axis
    plt.plot(x, y, 'ro')
    print(" ")
    plt.grid() # displays grid lines on graph
    plt.show()
    guess = input("Insira as coordenadas do ponto vermelho: \n")
    guess_array = guess.split(",")
    xguess = int(guess_array[0])
    yguess = int(guess_array[1])
    if xguess == xpoint and yguess == ypoint:
        score = score + 1

print("Sua pontuação: ", score) # notice this is not in the loop
```

### Jogo de prática de álgebra
- Gere problemas de uma etapa e de duas etapas com valores inteiros aleatórios e o jogador tem que inserir a resposta
- Usar valores positivos e negativos. Para aumentar a dificuldade, aumente os números



```python
# converts string input (even fractions) to float
def string_frac(in_string):
    if "/" in in_string:
        nd = in_string.split("/")
        n = float(nd[0])
        d = float(nd[1])
        ans = n/d
        return ans
    else:
        ans = float(in_string)
        return ans


# Simplest one-step addition
def one_step_add():
    import random
    # Display problem
    a = random.randint(-4,10)
    b = random.randint(2,24)
    print("x + ", a, " = ", b)
    ans = float(input("x = "))
    answer = b-a
    # Test input
    if ans==answer:
        print("Correto! \n")
    else:
        print("Tente novamente!")
        print("A resposta correta é: ", answer, "\n")


# One-step additon with negaive numbers
def one_step_subtract():
    import random
    a = random.randint(-19,-1)
    b = random.randint(2,24)
    print(a, " + x = ", b)
    ans = float(input("x = "))
    # test
    answer = b-a
    if ans==answer:
        print("Correto! \n")
    else:
        print("Tente novamente!")
        print("A resposta correta é: ", answer, "\n")

# One-step multiply
def one_step_mult():
    # Uses string_frac(<input string>)
    import random
    a = random.randint(1,11)
    b = random.randint(2,24)
    print(a, "x = ", b)
    print("Obs: Arredonde a resposta para duas casas decimais!")
    ans_in = (input("x = "))
    answer = round(b/a,2)
    # test
    if string_frac(ans_in)==answer:
        print("Correto! \n")
    else:
        print("Tente novamente!")
        print("A resposta correta é: ", answer, "\n")


# One-step divide
def one_step_div():
    import random
    a = random.randint(1,11)
    b = random.randint(2,24)
    print("x/", a, " = ", b)
    ans = float(input("x = "))
    answer = b*a
    # test
    if ans==answer:
        print("Correto! \n")
    else:
        print("Tente novamente!")
        print("A resposta correta é: ", answer, "\n")


# Two-step problems
def two_step():
    import random
    # Uses string_frac()
    a = random.randint(1,11)
    b = random.randint(-7,12)
    c = random.randint(2,36)
    print(a, "x + ", b, " = ", c)
    print("Obs: Arredonde a resposta para duas casas decimais!")
    ans_in = input("x = ")
    answer = (c-b)/a
    # test
    if round(string_frac(ans_in),2)==round(answer,2):
        print("Correto! \n")
    else:
        print("Tente novamente!")
        print("A resposta correta é: ", answer, "\n")

def main():
    # Lista de funções
    funcoes = [
        one_step_add,
        one_step_subtract,
        one_step_mult,
        one_step_div,
        two_step,
        ]

    # Número de funções a serem chamadas
    num_funcoes = int(input("Digite o número de questões (2-5): "))
    if num_funcoes not in range(2, 6):
        print("Número inválido. Digite um número entre 2 e 5.")
        return

    # Gerar indices aleatórios
    indices = random.sample(range(len(funcoes)), num_funcoes)

    # Chamar as funções aleatórias
    for indice in indices:
        funcoes[indice]()

if __name__ == "__main__":
    main()

```

### Jogo de projéteis
- Exiba um "parede" com altura e local aleatórios. O jogador precisa mover os controles deslizantes para ajustar o caminho da parábola para evitar a parede
- Para adicionar dificuldade, faça um segundo nível onde jogadores inserem a, b e c sem os controles deslizantes

```python
# Write your code here
import numpy as np
import matplotlib.pyplot as plt

# Definindo as condições iniciais
x0 = 0
y0 = 0
v0_x = int(input("Defina a velocidade: "))  # Velocidade inicial x
v0_y = int(input("Defina a altura: "))    # Velocidade inicial y
g = -9.8

# Gerando propriedades aleatórias da parede
wall_height = np.random.randint(5, 15)
wall_position = np.random.randint(20, 40)

# Definindo o passo de tempo e o tempo total
dt = 0.01
t = np.arange(0, 5, dt)

# Calculando o movimento do projétil
x = x0 + v0_x * t
y = y0 + v0_y * t + 0.5 * g * t**2

# Verificando se o foguete passa pela parede
max_height = y.max()
success = max_height > wall_height

# Imprimindo resultados
print(f"Velocidade inicial: ({v0_x}, {v0_y})")
print(f"Altura da parede: {wall_height}")
print(f"Posição da parede: {wall_position}")
print(f"Sucesso: {success}")

# Plotando a trajetória (bônus)
plt.plot(x, y)
plt.axhline(wall_height, color='red', linestyle='--')
plt.axvline(wall_position, color='green', linestyle='--')
plt.xlabel("Posição X")
plt.ylabel("Posição Y")
plt.title("Trajetória do Projétil")
plt.show()


# This step does not have a test
```
