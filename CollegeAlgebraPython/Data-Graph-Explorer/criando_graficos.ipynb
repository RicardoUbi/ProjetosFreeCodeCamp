```python
import pandas as pd

def add_arquivo():
  # Escolha do método de obtenção do arquivo
  metodo = input("Digite o método de obtenção do arquivo (local, url): ")

  # Obtendo o arquivo por método
  if metodo == "local":
    arquivo = input("Digite o caminho do arquivo: ")
    df = pd.read_csv(arquivo)
  elif metodo == "url":
    url = input("Digite a URL do arquivo: ")
    df = pd.read_csv(url)
  else:
    print("Método inválido!")
    return None

  return df

# Obtendo o DataFrame
df = add_arquivo()

# Imprimindo cabeçalhos e primeiras linhas
print(df.columns)
print(df.head(2))

# Armazenando nomes das colunas
colunas = list(df.columns)

# Escolha das colunas para análise
coluna_x = input("Digite o nome da coluna X: ")
coluna_y = input("Digite o nome da coluna Y: ")

# Convertendo dados em arrays do NumPy
dados_x = df[coluna_x].to_numpy()
dados_y = df[coluna_y].to_numpy()

# Exibindo gráfico de dispersão ou linear
tipo_grafico = input("Digite o tipo de gráfico (dispersão, linear): ")

if tipo_grafico == "dispersão":
  import matplotlib.pyplot as plt
  plt.scatter(dados_x, dados_y)
  plt.xlabel(coluna_x)
  plt.ylabel(coluna_y)
  plt.show()
elif tipo_grafico == "linear":
  import seaborn as sns
  sns.regplot(x=dados_x, y=dados_y)
  plt.show()
else:
  print("Tipo de gráfico inválido!")
```
